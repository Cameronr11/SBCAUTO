```python
#intial bert setup
!pip install transformers torch
from transformers import BertTokenizer, BertModel, pipeline
import torch

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained("bert-base-uncased")

#example sentence to test setup
sentence = "Sherlock holmes is a great detective"
tokens = tokenizer(sentence, return_tensors="pt")

with torch.no_grad():
    outputs = model(**tokens)

print("Tokenized input:", tokens["input_ids"])
print("model output:", outputs.last_hidden_state)
```

```python
#here all I am doing is trying to test how good bert is at answering questions out of the box already before we fine tune

qa_pipeline = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad")

# Context for A Study in Scarlet
context_study_in_scarlet = """
Dr. John Watson, recently returned from Afghanistan, is introduced to Sherlock Holmes by Stamford.
They decide to share an apartment at 221B Baker Street. Holmes demonstrates his deductive skills and assists the police
in solving the murder of Enoch Drebber. The word 'RACHE' is written in blood at the crime scene. Jefferson Hope is later
revealed as the murderer, seeking revenge for past wrongs.
"""

# Context for The Hound of the Baskervilles
context_hound_of_baskervilles = """
Dr. Watson and Sherlock Holmes investigate the Baskerville curse after the mysterious death of Sir Charles Baskerville.
Sir Henry Baskerville, the new heir, arrives in London and moves to Baskerville Hall. Holmes discovers that Jack Stapleton,
a relative of the Baskervilles, is behind the scheme, intending to inherit the family estate. A legendary hound is revealed
to be part of Stapleton’s plan.
"""

# Questions for A Study in Scarlet
questions_study_in_scarlet = [
    "Who introduced Dr. John Watson to Sherlock Holmes?",
    "Where do Watson and Holmes decide to live?",
    "Who was murdered in the Lauriston Gardens?",
    "What word was written on the wall at the crime scene?",
    "Who was revealed as the murderer?",
    "What was Jefferson Hope’s motive for murder?"
]

# Questions for The Hound of the Baskervilles
questions_hound_of_baskervilles = [
    "Who died mysteriously at the beginning of the story?",
    "What is the name of the family’s estate?",
    "Who is the new heir to the Baskerville fortune?",
    "What does Dr. Mortimer believe about the Baskerville curse?",
    "Who is revealed as the perpetrator in the story?",
    "What was Jack Stapleton’s motive for his actions?",
    "What is the significance of the hound in the plot?"
]

# Function to run QA and print results
def run_qa(context, questions):
    print("\nTesting BERT Out-of-the-Box on Sherlock Holmes Stories\n")
    for i, question in enumerate(questions, 1):
        result = qa_pipeline(question=question, context=context)
        print(f"Q{i}: {question}")
        print(f"Answer: {result['answer']}")
        print(f"Score: {result['score']:.4f}")
        print("-" * 50)

# Test BERT on A Study in Scarlet
print("Testing on A Study in Scarlet\n")
run_qa(context_study_in_scarlet, questions_study_in_scarlet)

# Test BERT on The Hound of the Baskervilles
print("\nTesting on The Hound of the Baskervilles\n")
run_qa(context_hound_of_baskervilles, questions_hound_of_baskervilles)
```
